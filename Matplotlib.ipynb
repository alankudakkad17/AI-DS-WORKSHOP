```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([100,200,300,400,500,600,700,800,900])
y=np.array([45,52,62,72,82,92,111,123,145])
plt.plot(x,y)
plt.title("A vs B")
plt.xlabel("A")
plt.ylabel("B")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([456,100,565,518,152,654,454,545,262])
y=np.array([45,52,62,72,82,92,111,123,145])
plt.scatter(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
y=np.array([35,25,25,45])
mylabels=["1st year","2nd year","3rd year","4th year"]
myexplode=[0,1,0,0]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show()
```
